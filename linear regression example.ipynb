```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

dataset = pd.read_csv('student_scores.csv')  
```

```python
dataset.shape  
```

```python
dataset.head()
```

```python
dataset.describe()  
```

```python
dataset.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()  
```

```python
x = dataset.iloc[:,0:1].values #np.arange(25).reshape(-1, 1) 
y = dataset.iloc[:,1].values#np.arange(25)
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2) 
print("training data",X_train)
print("testing data",X_test)
print("training data y",y_train)
print("testing data y",y_test)
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train)  
```

```python
print(regressor.intercept_)  #c(intercept)
```

```python
print(regressor.coef_)  #m(slope)
```

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df  
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred))) 
print('r_2 statistic: %.2f' % r2_score(y_test, y_pred))
#MSE(Mean Squared Error) formula#
# step1 Subtract the predicted Y value from the actual Y value to get the error.
# step2 Square the errors.
# step3 Add up the errors.
# step4 Find the mean.#
#R squared formula#
#SS_res =  sum(square(y_true - y_pred)) 
#SS_tot = sum(square(y_true - np.mean(y_true))) 
#R squared(r2_score)= 1 - (SS_res/SS_tot )#
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
```
